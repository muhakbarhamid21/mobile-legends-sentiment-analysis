```python
import pandas as pd
import time
from google_play_scraper import reviews, Sort, app
```

```python
app_id = 'com.mobile.legends'
```

```python
app_info = app(app_id)
```

```python
app_info
```

```python
target_review_count = 30000
batch_size = 100
all_reviews = []
continuation_token = None
```

```python
while len(all_reviews) < target_review_count:
    result, continuation_token = reviews(
        app_id,
        lang='id',
        country='id',
        sort=Sort.MOST_RELEVANT,
        count=batch_size,
        continuation_token=continuation_token
    )
    
    if not result: 
        print("Tidak ada lagi ulasan yang tersedia.")
        break
        
    all_reviews.extend(result)
    print(f"Sudah terkumpul {len(all_reviews)} ulasan...")
    
    time.sleep(1)
    
    if continuation_token is None:
        print("Semua ulasan telah diambil.")
        break

all_reviews = all_reviews[:target_review_count]
```

```python
df = pd.DataFrame(all_reviews)
```

```python
df.to_csv('../datasets/mlbb_reviews2.csv', index=False)
```
